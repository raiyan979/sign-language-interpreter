```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import time
import mediapipe as mp
```

```python
cap = cv2.videoCapture(0)
while cap.IsOpen():
    ret, frame = cap.read()
    cv2.imshow("OpenCV feed", frame)
    if cv2.waitkey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
